```python
import numpy as np
import pandas as pd
```

```python
#importing dataset
df = pd.read_csv('/content/final_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Group the data by 'Company_Name' and sum the LinkedIn followers for each company
company_followers = df.groupby('Company_Name')['LinkedIn_Followers'].sum().reset_index()

# Find the company with the most followers
company_with_most_followers = company_followers.loc[company_followers['LinkedIn_Followers'].idxmax()]

# Display the DataFrame showing followers for each company
print("LinkedIn Followers Count by Company:")
print(company_followers)

# Display the company with the most followers
print("\nCompany with the Most Followers:")
print(company_with_most_followers)
```

```python
# Visualize the data using a bar chart
plt.figure(figsize=(12, 6))
plt.barh(company_followers['Company_Name'], company_followers['LinkedIn_Followers'], color='skyblue')
plt.xlabel('LinkedIn Followers Count')
plt.ylabel('Company Name')
plt.title('LinkedIn Followers Count by Company')
plt.gca().invert_yaxis()  # Invert the y-axis to display the company with the most followers at the top
plt.tight_layout()

# Show the plot
plt.show()

```

```python
#What are the most demand jobs that companies are hiring
# Count the occurrences of each job role and sort them in descending order
job_demand = df['Designation'].value_counts().reset_index()
job_demand.columns = ['Job_Title', 'Count']
job_demand = job_demand.sort_values(by='Count', ascending=False)

# Display the top-demand job roles
print("Top-Demand Job Roles:")
print(job_demand.head(10))  # Display the top 10 job roles
```

```python
#demand skills that most companies are hiring

# Define a list of skill columns to analyze
skills_columns = [
    'PYTHON', 'JAVA', 'SQL', 'TENSORFLOW', 'REACT', 'C++', 'HADOOP',
    'SCALA', 'FLASK', 'PANDAS', 'SPARK', 'NUMPY', 'PHP', 'MYSQL',
    'CSS', 'MONGODB', 'NLTK', 'LINUX', 'RUBY', 'JAVASCRIPT', 'DJANGO',
    'REACTJS', 'AI', 'UI', 'TABLEAU', 'NODEJS', 'EXCEL', 'POWER BI',
    'SELENIUM', 'HTML', 'ML'
]

# Calculate the total demand for each skill
skill_demand = df[skills_columns].sum()

# Sort skills by demand in descending order
sorted_skills = skill_demand.sort_values(ascending=False)

# Create a bar plot to visualize the most demanding skills
plt.figure(figsize=(12, 8))
sns.barplot(x=sorted_skills.values, y=sorted_skills.index, palette="viridis")
plt.xlabel('Number of Job Postings')
plt.ylabel('Skills')
plt.title('Most Demanding Skills in Tech Job Postings')
plt.tight_layout()

# Show the plot
plt.show()


```

```python
#total applicants based on company and designation

# Group the data by 'Company_Name' and 'Designation', and sum the 'Total_applicants' column
applicant_counts = df.groupby(['Company_Name', 'Designation'])['Total_applicants'].sum().reset_index()

# Display the total applicants for each combination of company and designation
print("Total Applicants by Company and Designation:")
print(applicant_counts)
applicant_counts.head(5)
```

```python
#most demand designation and their count
# Calculate the total demand for each job designation
designation_demand = df['Designation'].value_counts().reset_index()
designation_demand.columns = ['Designation', 'Count']

# Find the most in-demand job designation
most_demand_designation = designation_demand.loc[designation_demand['Count'].idxmax()]

# Display the most in-demand job designation
print("Most In-Demand Job Designation:")
print(most_demand_designation)

```

```python
# Calculate the total demand for each job designation
designation_demand = df['Designation'].value_counts().reset_index()
designation_demand.columns = ['Designation', 'Count']

# Find the top 10 most in-demand job designations
top_10_demand_designations = designation_demand.head(10)

# Display the top 10 most in-demand job designations
print("Top 10 Most In-Demand Job Designations:")
print(top_10_demand_designations)

# Visualize the top 10 most in-demand job designations using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Designation', data=top_10_demand_designations, palette='viridis')
plt.xlabel('Count')
plt.ylabel('Job Designation')
plt.title('Top 10 Most In-Demand Job Designations')
plt.tight_layout()

# Show the plot
plt.show()




```

```python

# Group the data by 'Location' and calculate the total applicants for each location
location_applicants = df.groupby('Location')['Total_applicants'].sum().reset_index()

# Find the location with the most preferred applicants
most_preferred_location = location_applicants.loc[location_applicants['Total_applicants'].idxmax()]

# Display the location with the most preferred applicants
print("Most Preferred Location:")
print(most_preferred_location)

# Visualize the total applicants by location using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Total_applicants', y='Location', data=location_applicants, palette='viridis')
plt.xlabel('Total Applicants')
plt.ylabel('Location')
plt.title('Preferred Location for Applicants (Total Applicants)')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group the data by 'Company_Name' and calculate the total employee count for each company
company_employee_count = pd.DataFrame(df.groupby('Company_Name')['Employee_count'].sum().reset_index())

# Display the total employee count for each company
print("Total Employee Count by Company:")
print(company_employee_count)
```

```python
# Group the data by 'Company_Name' and 'Designation' and sum the 'Total_applicants' column
applicant_counts = df.groupby(['Company_Name', 'Designation'])['Total_applicants'].sum().reset_index()

# Find the designation with the most applicants for each company
most_vacant_designation = applicant_counts.loc[applicant_counts.groupby('Company_Name')['Total_applicants'].idxmax()]

# Display the designation with the most applicants for each company
print("Designation with Most Applicants for Each Company:")
print(most_vacant_designation)
```
